```python
#1. Word Tokenization and Frequency Distribution using NLTK

# Install NLTK
!pip install nltk

import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

# Sample text
text = "Natural Language Processing with Python is exciting and useful."

# Tokenization
nltk.download('punkt')
tokens = word_tokenize(text)

# Frequency Distribution
fdist = FreqDist(tokens)

# Plotting
fdist.plot(30, cumulative=False)
plt.show()
```

```python
#2. Dependency Parsing with SpaCy

# Install SpaCy
!pip install spacy

import spacy

# Download English model
!python -m spacy download en_core_web_sm

# Load SpaCy English model
nlp = spacy.load("en_core_web_sm")

# Sample sentence
sentence = "John loves programming in Python."

# Parse dependency
doc = nlp(sentence)

for token in doc:
    print(f"{token.text} --> {token.dep_} ({token.head.text})")
```

```python
#3. Text Classification using TextBlob

# Install TextBlob
!pip install textblob

from textblob import TextBlob

# Sample text
text = "This product is amazing and worth every penny."

# Sentiment analysis
blob = TextBlob(text)
print(f"Polarity: {blob.sentiment.polarity}, Subjectivity: {blob.sentiment.subjectivity}")
```

```python
#4. Named Entity Recognition (NER) with SpaCy

# Ensure SpaCy and English model are installed
!pip install spacy
!python -m spacy download en_core_web_sm

import spacy

nlp = spacy.load("en_core_web_sm")
text = "Apple Inc. is looking to buy a UK-based startup for $1 billion."

doc = nlp(text)

# Extract named entities
for ent in doc.ents:
    print(f"{ent.text}: {ent.label_}")
```

```python
#5. TF-IDF Calculation using Scikit-learn

# Install Scikit-learn
!pip install scikit-learn

from sklearn.feature_extraction.text import TfidfVectorizer

# Sample documents
documents = ["This is a sample document.", "Natural Language Processing is fun.", "We love programming in Python."]

# TF-IDF Vectorization
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

# Display TF-IDF scores
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
#6. Custom Text Classifier with NLTK's Naive Bayes

# Ensure NLTK is installed
!pip install nltk

import nltk
from nltk.classify import NaiveBayesClassifier
from nltk.tokenize import word_tokenize

nltk.download('punkt')

# Sample training data
training_data = [("This is a great movie", "pos"), ("I hated this film", "neg")]

# Preprocess training data
def extract_features(text):
    words = set(word_tokenize(text))
    return {word: True for word in words}

training_set = [(extract_features(text), label) for (text, label) in training_data]

# Train Naive Bayes Classifier
classifier = NaiveBayesClassifier.train(training_set)
test_sentence = "I love this movie"
print(classifier.classify(extract_features(test_sentence)))
```

```python
#9. Simple RNN for Text Classification using Keras

# Install TensorFlow
!pip install tensorflow

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Embedding

# Parameters
vocab_size = 1000
embedding_dim = 16
sequence_length = 100

# Model Definition
model = Sequential([
    Embedding(vocab_size, embedding_dim, input_length=sequence_length),
    SimpleRNN(32),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
#10. Training a Bi-directional LSTM

!pip install tensorflow

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense, Embedding

vocab_size = 1000
embedding_dim = 16
sequence_length = 100

model = Sequential([
    Embedding(vocab_size, embedding_dim, input_length=sequence_length),
    Bidirectional(LSTM(32)),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
#11. Implementing GRU for Text Classification

!pip install tensorflow

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Embedding

vocab_size = 1000
embedding_dim = 16
sequence_length = 100

model = Sequential([
    Embedding(vocab_size, embedding_dim, input_length=sequence_length),
    GRU(32),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
#12. Text Generation using LSTM

!pip install tensorflow

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding

vocab_size = 1000
embedding_dim = 16
sequence_length = 100

model = Sequential([
    Embedding(vocab_size, embedding_dim, input_length=sequence_length),
    LSTM(64, return_sequences=True),
    LSTM(64),
    Dense(vocab_size, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy')
model.summary()
```

```python
#13. Bi-directional GRU for Sequence Labeling

!pip install tensorflow

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, GRU, Dense, Embedding

vocab_size = 1000
embedding_dim = 16
sequence_length = 100

model = Sequential([
    Embedding(vocab_size, embedding_dim, input_length=sequence_length),
    Bidirectional(GRU(32, return_sequences=True)),
    Dense(5, activation='softmax')  # Assuming 5 classes for labeling
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```
